```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
#Importing data
df = pd.read_csv('dataset.csv')
#Printing head
df.head()
```

```python
df.tail()
```

```python
df = pd.read_csv('dataset.csv', nrows = 762)
train=df[0:731] 
test=df[731:]
df.Timestamp = pd.to_datetime(df.date,format='%Y-%m-%d') 
df.index = df.Timestamp 

df = df.resample('D').mean()
train.Timestamp = pd.to_datetime(train.date,format='%Y-%m-%d') 
train.index = train.Timestamp 
train = train.resample('D').mean() 
test.Timestamp = pd.to_datetime(test.date,format='%Y-%m-%d') 
test.index = test.Timestamp 
test = test.resample('D').mean()
print(train)
```

```python
train.volume.plot(figsize=(15,8), title= 'Daily Ridership', fontsize=14)
test.volume.plot()
plt.show()
```

```python
dd= np.asarray(train.volume)
y_hat = test.copy()
y_hat['naive'] = dd[len(dd)-1]
plt.figure(figsize=(12,8))
plt.plot(train.index, train['volume'], label='Train')
plt.plot(test.index,test['volume'], label='Test')
plt.plot(y_hat.index,y_hat['naive'], label='Naive Forecast')
plt.legend(loc='best')
plt.title("Naive Forecast")
plt.show()
from sklearn.metrics import mean_squared_error
from math import sqrt
rms = sqrt(mean_squared_error(test.volume, y_hat.naive))
print(rms)
```

```python
y_hat_avg = test.copy()
y_hat_avg['avg_forecast'] = train['volume'].mean()
plt.figure(figsize=(12,8))
plt.plot(train['volume'], label='Train')
plt.plot(test['volume'], label='Test')
plt.plot(y_hat_avg['avg_forecast'], label='Average Forecast')
plt.legend(loc='best')
plt.show()
rms = sqrt(mean_squared_error(test.volume, y_hat_avg.avg_forecast))
print(rms)
```

```python
y_hat_avg = test.copy()
y_hat_avg['moving_avg_forecast'] = train['volume'].rolling(60).mean().iloc[-1]
plt.figure(figsize=(16,8))
plt.plot(train['volume'], label='Train')
plt.plot(test['volume'], label='Test')
plt.plot(y_hat_avg['moving_avg_forecast'], label='Moving Average Forecast')
plt.legend(loc='best')
plt.show()
rms = sqrt(mean_squared_error(test.volume, y_hat_avg.moving_avg_forecast))
print(rms)
```

```python
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
y_hat_avg = test.copy()
fit2 = SimpleExpSmoothing(np.asarray(train['volume'])).fit(smoothing_level=0.6,optimized=False)
y_hat_avg['SES'] = fit2.forecast(len(test))
plt.figure(figsize=(16,8))
plt.plot(train['volume'], label='Train')
plt.plot(test['volume'], label='Test')
plt.plot(y_hat_avg['SES'], label='SES')
plt.legend(loc='best')
plt.show()
rms = sqrt(mean_squared_error(test.volume, y_hat_avg.SES))
print(rms)
```

```python
import statsmodels.api as sm
sm.tsa.seasonal_decompose(train.volume).plot()
result = sm.tsa.stattools.adfuller(train.volume)
plt.show()
```

```python
y_hat_avg = test.copy()

fit1 = Holt(np.asarray(train['volume'])).fit(smoothing_level = 0.1,smoothing_slope = 0.1)
y_hat_avg['Holt_linear'] = fit1.forecast(len(test))

plt.figure(figsize=(16,8))
plt.plot(train['volume'], label='Train')
plt.plot(test['volume'], label='Test')
plt.plot(y_hat_avg['Holt_linear'], label='Holt_linear')
plt.legend(loc='best')
plt.show()

rms = sqrt(mean_squared_error(test.volume, y_hat_avg.Holt_linear))
print(rms)
```

```python
y_hat_avg = test.copy()
fit1 = ExponentialSmoothing(np.asarray(train['volume']) ,seasonal_periods=7 ,trend='add', seasonal='add',).fit()
y_hat_avg['Holt_Winter'] = fit1.forecast(len(test))
plt.figure(figsize=(16,8))
plt.plot( train['volume'], label='Train')
plt.plot(test['volume'], label='Test')
plt.plot(y_hat_avg['Holt_Winter'], label='Holt_Winter')
plt.legend(loc='best')
plt.show()
rms = sqrt(mean_squared_error(test.volume, y_hat_avg.Holt_Winter))
print(rms)
```

```python
y_hat_avg = test.copy()
fit1 = sm.tsa.statespace.SARIMAX(train.volume, order=(0, 1, 1),seasonal_order=(1,1,1,7)).fit()
y_hat_avg['SARIMA'] = fit1.predict(start="2018-01-01", end="2018-01-31", dynamic=True)
plt.figure(figsize=(16,8))
plt.plot( train['volume'], label='Train')
plt.plot(test['volume'], label='Test')
plt.plot(y_hat_avg['SARIMA'], label='SARIMA')
plt.legend(loc='best')
plt.show()
rms = sqrt(mean_squared_error(test.volume, y_hat_avg.SARIMA))
print(rms)
```
